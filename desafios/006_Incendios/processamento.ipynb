```python
import pandas as pd
import geopandas as gpd

df1 = pd.read_csv('All_data_1980_2021_grouped_centroids_denorm.csv')
```

```python
df1
```

```python

parishes = gpd.read_file('Cont_AAD_CAOP2017.shp', SHAPE_RESTORE_SHX='YES')
```

```python
parishes.to_crs(epsg=4326, inplace=True)
```

```python
geopandas_df = gpd.GeoDataFrame(
    df1, geometry=gpd.points_from_xy(df1.x, df1.y))
```

```python
geopandas_df.crs = "epsg:4326"
```

```python
geopandas_df.head(3)
```

```python
parishes.head(3)
```

```python
merged = geopandas_df.sjoin(parishes, how="left")
```

```python
merged.head(3)
```

```python
output = merged[['Dicofre', 'Freguesia', 'Distrito', 'concelho', 'month', 'year', 'sum', 'x', 'y']]
```

```python
output = output.rename(columns={'x': 'longitude', 'y': 'latitude'})
```

```python
output.head(3)
```

```python
output.to_csv('area_ardida.csv')
```
